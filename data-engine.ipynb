## Dependencies

```python
import pandas as pd
import duckdb
```

## Base data

```python
df = pd.read_csv("/Users/andrebezerra/Library/CloudStorage/GoogleDrive-andre.oliveira.bezerra@gmail.com/My Drive/pokemon-data-analysis/analysis/pokemon_analysis.csv")
```

## DuckDB

### Create database

```python
conn = duckdb.connect("pokemon_data.duckdb")
```

### Create table

```python
conn.execute("""
    CREATE TABLE IF NOT EXISTS pokemon ( 
      name TEXT,
      id INT,
      type_1 TEXT,
      type_2 TEXT,
      type_unique TEXT,
      attack INT,
      defense INT,
      hp INT,
      sp_attack INT,
      sp_defense INT,
      speed INT,
      best_stat TEXT,
      Final_form TEXT
    )
""")
```

### Insert data into table

```python
conn.execute("""
INSERT INTO pokemon
SELECT * FROM df
""")
```

### Query data via DuckDB SQL

```python
result = conn.execute("""
                      SELECT 
                      * 
                      FROM pokemon 
                      WHERE type_1 = 'fire'
                      LIMIT 20
                      """).df()

display(result)
```
